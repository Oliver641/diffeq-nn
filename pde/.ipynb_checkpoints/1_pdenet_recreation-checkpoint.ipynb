# Producing Diffusion Data

As in the pde-net, we will be examining solutions to diffusion equations of the form:

\begin{equation}
\left\{\begin{array}{cc}
u_t = a(x,y) u_x + b(x,y)u_y + c(x,y)u_{xx} + d(x,y)u_{yy} & \text{for $t,x,y \in [0, 0.2] \times \Omega$}\\
u\vert_{t=0} = u_0(x,y) & \text{ }
\end{array}
\right.
\end{equation}

Where $\Omega = [0, 2\pi] \times [0, 2\pi]$.  In this paper in particular, they took $c \equiv .2, d\equiv .3$ and

\begin{equation}
a(x,y) = 0.5(\cos(y) + x(2\pi - x)\sin(x)) + 0.6\\ \\
b(x,y) = 2(\cos(y) + \sin(x)) + 0.8
\end{equation}

The initial condition $u_0$ is generated from

\begin{equation}
u_0(x,y) = \sum_{\vert k\vert, \vert l \vert < N} \lambda_{k,l} \cos(kx+ly) + \gamma_{k,l} \sin(kx+ly)
\end{equation}

where $\lambda_{k,l}, \gamma_{k,l} \sim N\left(0, \frac{1}{50}\right)$ and $k,l$ are 'chosen randomly'. (What does this mean?)

```python
def get_equation(a, b, c, d):
    return (TransientTerm()
            == DiffusionTerm([c.flatten(),d.flatten()])
            + FirstOrderAdvectionTerm([a.flatten(),b.flatten()]))
```

```python
def generate_initial(x, y, N=9, variance=1/50):
    lambda_kl = variance * np.random.randn(N,N)
    gamma_kl = variance * np.random.randn(N,N)
    return sum(lambda_kl[k,l] * np.cos(k*x + l*y) 
                  + gamma_kl[k,l] * np.sin(k*x+l*y) 
                  for k in range(N) 
                  for l in range(N))
```

```python
nx,ny = 50,50
dx,dy = 2*np.pi/nx, 2*np.pi/ny
t=.2
time_step=.01
num_timesteps = int(np.floor(t/time_step))
xx, yy = np.meshgrid(np.linspace(0, 2*np.pi, nx), np.linspace(0, 2*np.pi, ny))
a = .5 * (np.cos(yy) + xx*(2*np.pi - xx) * np.sin(xx)) + .6
b = 2 * (np.cos(yy) + np.sin(xx)) + .8
c = .2 * np.ones((nx,ny))
d = .3 * np.ones((nx,ny))
u_0 = generate_initial(xx,yy)



    
def du_dt(t, u):
    u = u.reshape(nx,ny)
    ux = [psdiff(u_0[i]) for i in range(len(u_0))]
    uxx = [psdiff(u_0[i], order=2) for i in range(len(u_0))]
    uy = [psdiff(u_0[:,i]) for i in range(len(u_0[0]))]
    uyy = [psdiff(u_0[:,i], order=2) for i in range(len(u_0[0]))]

    a = .5 * (np.cos(yy) + xx*(2*np.pi - xx) * np.sin(xx)) + .6
    b = 2 * (np.cos(yy) + np.sin(xx)) + .8
    c = .2 * np.ones((nx,ny))
    d = .3 * np.ones((nx,ny))
    return (a * ux + b * uy + c * uxx + d * uyy).flatten()
```

```python
def generate_diffusion(nx, ny, du_dt, t=.2, step_size=.01):
    dx,dy = 2*np.pi/nx, 2*np.pi/ny
    xx, yy = np.meshgrid(np.linspace(0, 2*np.pi, nx), np.linspace(0, 2*np.pi, ny))
    
    u_0 = generate_initial(xx,yy)
    
    solution = sp.integrate.RK45(du_dt, 0, u_0.flatten(), t, max_step=step_size)
    data = np.array([solution.y])
    for i in range(int(t/step_size)):
        solution.step()
        data = np.append(data, np.array([solution.y]), axis=0)
    return data
```

```python
data = generate_diffusion(nx, ny, du_dt, t=t, step_size=time_step)
```

```python
def validate_solution(data, du_dt, nx, ny, time_step, tol=.000001):
    for i in range(1,len(data) - 1):
        err = data[i+1] - data[i] - (time_step * du_dt(i, data[i])
                                     + time_step * du_dt(i+1, data[i+1]))/2
        if np.max(np.abs(err)) > tol:
            return False
    return True
```

```python
validate_solution(data, du_dt, nx, ny, time_step)
```

```python
def diffusion_data_generator(nx, ny, t, step_size):
    while True:
        x = generate_diffusion(nx, ny, du_dt, t, step_size)
        yield x[0].astype(np.float32), x[1:].astype(np.float32)
```

```python
dataset = tf.data.Dataset.from_generator(diffusion_data_generator,
                                         args=[nx, ny, t, time_step], 
                                         output_types=(tf.float32, tf.float32), 
                                         output_shapes=((nx*ny), (num_timesteps, nx*ny)))
```

# Equation Block

```python
class NonPos(tf.keras.constraints.Constraint):
    """Constrains the weights incident to each hidden unit to Negative.
    """

    def __init__(self):
        pass

    def __call__(self, w):
        w = w * K.cast(K.less_equal(w, 0.), K.floatx())
        return w
```

```python
def wrap_padding(x, padding=1):
    for i, p in enumerate(padding):
        if p[0] > 0 and p[1] > 0:
            x = tf.concat([tf.gather(x, list(range(x.shape[i] - p[0], x.shape[i])), axis=i), 
                          x, 
                          tf.gather(x, list(range(p[1])), axis=i)], axis=i)
        elif p[0] > 0:
            x = tf.concat([tf.gather(x, list(range(x.shape[i] - p[0], x.shape[i])), axis=i), 
                          x], axis=i)
        elif p[1] > 0:
            x = tf.concat([x, 
                          tf.gather(x, list(range(p[1])), axis=i)], axis=i)
        else:
            pass
        
    return x

WrapPadding = lambda padding: tf.keras.layers.Lambda(lambda x: wrap_padding(x, padding))
```

```python
class EquationBlock2D(tf.keras.layers.Conv2D):

    def __init__(self, 
                 wavelet,
                 filters,
                 strides=1,
                 
                 padding='valid',
                 data_format='channels_last',
                 dilation_rate=1,
                 activation=None,
                 activity_regularizer=None,
                 functional_initializer='glorot_uniform',
                 functional_regularizer=None,
                 trainable=True,
                 **kwargs):
        self.wavelet = wavelet
        self.functional_initializer = functional_initializer
        self.functional_regularizer = functional_regularizer
        self.trainable = trainable
        
        assert self.wavelet.shape[0] == self.wavelet.shape[1]
        kernel_size = self.wavelet.shape[0]
        super(EquationBlock2D, self).__init__(
            filters=filters,
            kernel_size=kernel_size,
            strides=strides,
            padding=padding,
            data_format=data_format,
            dilation_rate=dilation_rate,
            activation=activation,
            use_bias=False,
            activity_regularizer=activity_regularizer,
            **kwargs)
        
    def build(self, input_shape):
        if self.data_format == 'channels_first':
            channel_axis = 1
        else:
            channel_axis = -1
        if input_shape[channel_axis] is None:
            raise ValueError('The channel dimension of the inputs '
                             'should be defined. Found `None`.')
            
        
        input_dim = input_shape[channel_axis]
        
        self.wrap_padding = WrapPadding(((0,0), (1,1), (1,1), (0,0)))
        self.kernel_size = self.kernel_size + (1,1)
        
        self.output_row = conv_utils.conv_output_length(input_shape[1],
                                                        self.kernel_size[0],
                                                        self.padding,
                                                        self.strides[0])+2
        self.output_col = conv_utils.conv_output_length(input_shape[2], 
                                                        self.kernel_size[1],
                                                        self.padding, 
                                                        self.strides[1])+2

        self.D = self.add_weight(shape=kernel_shape,
                                 initializer=tf.keras.initializers.Constant(self.wavelet),
                                 name='D',
                                 trainable=False)
        self.c = self.add_weight(shape=(self.output_row * self.output_col, 1, 1, self.filters),
                                 initializer=self.functional_initializer,
                                 name='c',
                                 regularizer=self.functional_regularizer,
                                 trainable=self.trainable)
        
        self.built = True
        
    def call(self, inputs):
        outputs = self.wrap_padding(inputs)
        outputs = K.conv2d(
            outputs,
            self.D,
            strides=self.strides[0],
            padding=self.padding,
            data_format=self.data_format,
            dilation_rate=self.dilation_rate[0]
        )
        outputs = K.local_conv2d(outputs, self.c, (1,1), self.strides, (self.output_row, self.output_col))
        
        if self.activation is not None:
            return self.activation(outputs)
        
        return outputs
```

```python
class DeltaT:
    
    def __init__(self, equation_blocks, units, equation_order=2):
        self.units=units
        self.state_size=np.prod(units)
        self.equation_blocks = equation_blocks
        self.equation_order = equation_order
        
    def call(self, x, states):
        state = K.reshape(states, (1,) + self.units + (1,))
        new_state = state + K.sum([D(state) for D in self.equation_blocks.values()], axis=0)
        return new_state, K.reshape(new_state, (1, self.state_size))
    
    def __call__(self, x, state):
        return self.call(x, state)
    
    def visualize(self):
        fig = plt.figure()

        for i in range(self.equation_order+1):
            for j in range(self.equation_order+1):
                if(i+j<=self.equation_order):
                    ax = plt.subplot2grid((5,5), (2*i,2*j))
                    ax.axis('off')
                    ax.imshow(self.equation_blocks[(i,j)].c[:,:,:,index]
                                  .numpy().reshape(4,4), 
                              norm=plt.Normalize(-1.,1.), cmap='viridis')
                    ax.title.set_text("c_{}{}".format(i,j))
        fig.show()
```

```python
d_x = np.array([-1,1]).reshape(2,1,1,1)
d_y = np.array([-1,1]).reshape(1,2,1,1)

def gradient_l2(param, filter_size):
    def _gradient_l2(weight_matrix):
        W = K.transpose(K.reshape(weight_matrix, (1,4,4,filter_size)))
        W_x = K.conv2d(W,d_x)
        W_y = K.conv2d(W,d_y)
        return param * K.sum([K.sum(K.square(W)),
                              K.sum(K.square(W_x)),
                              K.sum(K.square(W_y))])
        
    return _gradient_l2
```

# PDE Model

```python
class PDENet(tf.keras.Model):
    
    def __init__(self, nx, ny, num_timesteps, filter_size):
        super(PDENet, self).__init__(name='pde_net')

        self.q = {
            (0,0): np.array([[1,1,1], [1,1,1], [1,1,1]])/9,
            (1,0): np.array([[1,2,1], [0,0,0], [-1,-2,-1]])/8,
            (0,1): (np.array([[1,2,1], [0,0,0], [-1,-2,-1]])/8).T,
            (2,0): np.array([[0,0,0], [1,-2,1], [0,0,0]]),
            (0,2): (np.array([[0,0,0], [1,-2,1], [0,0,0]])).T,
            (1,1): np.array([[1, 0, -1], [0,0,0], [-1,0,1]])/2
        }

        self.kernel_initializers = {
            key: tf.keras.initializers.Constant(value)
            for key, value in self.q.items()
        }
        
        self.flatten = tf.keras.layers.Flatten()
        self.repeat = tf.keras.layers.RepeatVector(num_timesteps)
        self.reshape = tf.keras.layers.Reshape((nx, ny))
        
        self.delta_t = DeltaT({key: EquationBlock2D(value, 
                                                    filter_size,
                                                    functional_regularizer=gradient_l2(.01, filter_size)) 
                               for key, value in self.q.items()}, (nx,ny))
        self.equation_layer = tf.keras.layers.RNN(self.delta_t, return_sequences=True)
        
    def call(self, input_tensor, training=False):
        x = self.flatten(input_tensor)
        #x = self.repeat(x)
        #x = self.reshape(x)
        output = self.equation_layer(tf.zeros((nx,ny,1)),x)
        return output
```

```python
filter_size=32
epochs=10
```

```python
model = PDENet(nx,ny,num_timesteps, filter_size)
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.mean_squared_error,
              metrics=[tf.keras.metrics.mean_squared_error])
```

```python
dataset = tf.data.Dataset.from_generator(diffusion_data_generator,
                                         args=[nx, ny, t, time_step], 
                                         output_types=(tf.float32, tf.float32), 
                                         output_shapes=((nx*ny), (num_timesteps, nx*ny)))
```

```python
h=model.fit(dataset.batch(32), epochs=epochs, steps_per_epoch=10)
```
